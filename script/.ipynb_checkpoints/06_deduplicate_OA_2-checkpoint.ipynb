```python
import pandas as pd
import os 
import glob

dir_path = "/Users/simon/Documents/repo/cities-learning"

# Create the full search pattern
file_pattern = os.path.join(dir_path, "data/OpenAlex/city_works_df_NA_abstr_added*.feather")

# Find all matching files
files = glob.glob(file_pattern)

# Read all the files into a list of DataFrames
dfs = [pd.read_feather(file) for file in files]

# Optionally, concatenate them into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Optionally concatenate them into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

print(combined_df.head())
```

```python
pre_count = len(combined_df)
print("observations before exact deduplication:" +  str())
combined_df = combined_df.drop_duplicates(subset=["id"])
duplicates_dropped = pre_count-len(combined_df)
print("exact deduplicates dropped:" + str(duplicates_dropped))
print("observations after exact deduplication:" + str(len(combined_df)))
print("Observations with missing abstracts after exact deduplication: " + str(combined_df["abstract"].isna().sum()))
print("Observations with missing titles after exact deduplication: " + str(combined_df["title"].isna().sum()))
print("Observations with missing titles after exact deduplication: " + str(combined_df["authors"].isna().sum()))
print("Observations with missing titles after exact deduplication: " + str(combined_df["publication_year"].isna().sum()))
```

```python
from semhash import SemHash

filtered_df = combined_df
# Replace empty strings and whitespace-only strings with None
filtered_df["title"] = filtered_df["title"].apply(lambda x: None if str(x).strip() == "" else x)
filtered_df["authors"] = filtered_df["authors"].apply(lambda x: None if str(x).strip() == "" else x)
filtered_df["abstract"] = filtered_df["abstract"].apply(lambda x: None if str(x).strip() == "" else x)
filtered_df["publication_year"] = filtered_df["publication_year"].apply(lambda x: None if str(x).strip() == "" else x)

# Check missing values after replacement
print("Observations with missing abstracts after exact deduplication: " + str(pd.DataFrame(filtered_df)["abstract"].isna().sum()))
print("Observations with missing titles after exact deduplication: " + str(pd.DataFrame(filtered_df)["title"].isna().sum()))
print("Observations with missing authors after exact deduplication: " + str(pd.DataFrame(filtered_df)["authors"].isna().sum()))
print("Observations with missing publication years after exact deduplication: " + str(pd.DataFrame(filtered_df)["publication_year"].isna().sum()))


# Convert to list of dictionaries and use 'id' as a key in each record
records = filtered_df[["id", "title", "abstract", "authors", "publication_year"]].to_dict(orient="records")
```

```python
# Define a helper function
def is_missing(value):
    return (
        value is None or                     # None
        (isinstance(value, float) and pd.isna(value)) or  # NaN
        (isinstance(value, str) and not value.strip())    # empty string or spaces
    )

# Now count missing for each field
missing_abstracts = sum(is_missing(rec.get("abstract")) for rec in records)
missing_titles = sum(is_missing(rec.get("title")) for rec in records)
missing_authors = sum(is_missing(rec.get("authors")) for rec in records)
missing_pub_year = sum(is_missing(rec.get("publication_year")) for rec in records)

print("Total observations:", len(records))
print("Observations with missing abstracts:", missing_abstracts)
print("Observations with missing titles:", missing_titles)
print("Observations with missing authors:", missing_authors)
print("Observations with missing publication years:", missing_pub_year)
```

```python
from semhash import SemHash

thresholds = [0.75, 0.80, 0.85, 0.90, 0.95, 1]
results = []

for t in thresholds:
    semhash = SemHash.from_records(records=records, columns=["title", "abstract", "authors"])
    dedup = semhash.self_deduplicate(threshold=t).deduplicated
    results.append((t, len(records), len(dedup), len(records) - len(dedup)))

# Print summary
print("\nThreshold | Original | Deduplicated | Removed")
for t, orig, dedup, removed in results:
    print(f"{t:<9} | {orig:<8} | {dedup:<13} | {removed}")
```

```python
import matplotlib.pyplot as plt

thresholds = [r[0] for r in results]
dedup_counts = [r[2] for r in results]

plt.plot(thresholds, dedup_counts, marker='o')
plt.title("Deduplicated Record Count vs. Similarity Threshold")
plt.xlabel("Similarity Threshold")
plt.ylabel("Remaining Records")
plt.grid(True)
plt.show()
```

```python
# publication years and titles are never missing
pd.DataFrame(records)["publication_year"].astype(int).isna().sum() == 0
pd.DataFrame(records)["title"].isna().sum() == 0

from sentence_transformers import SentenceTransformer
from semhash import SemHash
import pandas as pd

# Define thresholds for different field combinations
THRESHOLDS = {
    ("title", "publication_year"): 0.97,  # Only title -> higher threshold
    ("title", "authors", "publication_year"): 0.90,
    ("title", "abstract", "publication_year"): 0.88,
    ("title", "abstract", "authors", "publication_year"): 0.85,  # More text -> lower threshold
}

def get_columns_for_record(record):
#    has_abstract = pd.notna(record.get("abstract"))
#    has_authors = pd.notna(record.get("authors"))
    abstract = record.get("abstract")
    authors = record.get("authors")
    
    has_abstract = bool(abstract and abstract.strip())
    has_authors = bool(authors and str(authors).strip())

    if not has_abstract and not has_authors:
        return ["title", "publication_year"]
    elif not has_abstract:
        return ["title", "authors", "publication_year"]
    elif not has_authors:
        return ["title", "abstract", "publication_year"]
    else:
        return ["title", "abstract", "authors", "publication_year"]


# Group records
grouped = {}
for rec in records:
    cols = tuple(get_columns_for_record(rec))
    grouped.setdefault(cols, []).append(rec)

# Load model
# model = SentenceTransformer("xlm-r-bert-base-nli-stsb-mean-tokens")

# Process each group
dedup_results = []
for cols, group in grouped.items():
    semhash = SemHash.from_records(records=group, columns=list(cols), 
                                   #model=model
                                  )
    threshold = THRESHOLDS[cols]  # pick the right threshold
    print(threshold)
    result = semhash.self_deduplicate(threshold=threshold)
    dedup_results.append(result)
```

```python
filtered_duplicates = []

for dup in dedup_result.duplicates:
    try:
        original_year = int(dup.record.get("publication_year"))
    except (TypeError, ValueError):
        continue  # Skip if original year is invalid

    restricted_year_matches = []
    for rec, score in dup.duplicates:
        try:
            year = int(rec.get("publication_year"))
        except (TypeError, ValueError):
            continue  # Skip invalid years

        if abs(year - original_year) <= 1:
            restricted_year_matches.append((rec, score))

    if restricted_year_matches:
        # Optional: update the duplicates to only include matching ones
        dup.duplicates = restricted_year_matches
        filtered_duplicates.append(dup)

# Number of rows before and after deduplication
original_count = len(filtered_df)
duplicates_count = len(dedup_result.duplicates)
deduplicated_count = len(dedup_result.deduplicated)
filtered_deduplicated_count = len(filtered_duplicates)

print(f"Inital number: {original_count}.")
print(f"Deduplicated count: {deduplicated_count}.")
print(f"Inital number of duplicates: {duplicates_count}.")
print(f"Duplicates after filtering: {filtered_deduplicated_count}.")
print(f"Presumably wrong duplicates filtered out through by only matching when they are published +-1 year from each other: {duplicates_count-filtered_deduplicated_count}.")

# Example: print the first 5 filtered
for dup in filtered_duplicates[:5]:
    print("\n=== DUPLICATE RECORD ===")
    print(f"Title     : {dup.record['title'][:80]}")
    print(f"Year      : {dup.record.get('publication_year', 'N/A')}")
    print("Matches:")
    for rec, score in dup.duplicates:
        print(f"  - {rec['title'][:80]} ({rec.get('publication_year', 'N/A')}) | Score: {score:.3f}")

# now store all deduplicated results in files with chunks of 20k obs. per file city_works_df_NA_abstr_added_dedup_1, ...., city_works_df_NA_abstr_added_dedup_XX
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

# Initialize containers
scores_by_type = defaultdict(list)

for dup in duplicates:
    if not dup.exact:
        for record, score in dup.duplicates:
            abstract = record.get("abstract")
            authors = record.get("authors")
            has_abstract = bool(abstract and abstract.strip())
            has_authors = bool(authors and str(authors).strip())

            if has_abstract and has_authors:
                key = "abstract + authors"
            elif has_abstract:
                key = "only abstract"
            elif has_authors:
                key = "only authors"
            else:
                key = "neither"

            scores_by_type[key].append(score)

# Plot
plt.figure(figsize=(12, 8))

colors = {
    "abstract + authors": "green",
    "only abstract": "blue",
    "only authors": "orange",
    "neither": "red"
}

for label, scores in scores_by_type.items():
    if scores:
        sns.kdeplot(scores, label=label, fill=True, color=colors.get(label))

plt.xlabel("Similarity Score")
plt.ylabel("Density")
plt.title("Similarity Score Distribution by Available Fields")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import math

chunk_size = 20000
total_records = len(deduplicated_df)
num_chunks = math.ceil(total_records / chunk_size)

for i in range(num_chunks):
    start = i * chunk_size
    end = min((i + 1) * chunk_size, total_records)
    chunk = deduplicated_df.iloc[start:end]
    
    filename = f"data/OpenAlex/city_works_df_NA_abstr_added_dedup_{i+1}.csv"
    chunk.to_csv(filename, index=False)
    print(f"Saved {filename} with records {start} to {end-1}")
```
