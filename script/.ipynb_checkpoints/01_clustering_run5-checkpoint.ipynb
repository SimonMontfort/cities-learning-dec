```python
import numpy as np
import keras
import keras_tuner as kt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from tensorflow.keras.regularizers import l2
import tensorflow as tf

def build_model(hp):
    input_dim = cities_clean_scaled.shape[1]

    # Latent dimension: keep it small for low-dim input
    latent_dim = hp.Int('latent_dim', min_value=2, max_value=5)

    # Tunable L2 regularization
    reg = l2(hp.Float('l2_reg', min_value=1e-6, max_value=1e-2, sampling='log'))

    # Input layer
    input_layer = Input(shape=(input_dim,))

    # Encoder: 1â2 layers with smaller size
    encoder = Dense(hp.Choice('encoder_1_units', values=[16, 32, 48]), activation='relu', kernel_regularizer=reg)(input_layer)
    encoder = Dense(hp.Choice('encoder_2_units', values=[16, 32]), activation='relu', kernel_regularizer=reg)(encoder)

    # Latent space
    latent_space = Dense(latent_dim, activation='relu')(encoder)

    # Decoder
    decoder = Dense(hp.Choice('decoder_1_units', values=[16, 32, 48]), activation='relu', kernel_regularizer=reg)(latent_space)
    decoder = Dense(hp.Choice('decoder_2_units', values=[16, 32]), activation='relu', kernel_regularizer=reg)(decoder)

    # Output layer
    output_layer = Dense(input_dim, activation='sigmoid')(decoder)

    # Build model
    model = Model(inputs=input_layer, outputs=output_layer)

    # Compile with Adam optimizer and tuned learning rate
    model.compile(optimizer=Adam(learning_rate=hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')), loss='mse')

    return model

# Instantiate the tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_loss',
    max_epochs=200,
    factor=3,
    directory='hyperband',
    project_name='DEC_model_tuning_run5'
)

# Set the search space for the hyperparameters and start the search
tuner.search(cities_clean_scaled, cities_clean_scaled, epochs=200, batch_size=32, validation_split=0.2)

# Get the best model after hyperparameter tuning
best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()
```

```python
from tensorflow.keras.layers import Layer
import tensorflow.keras.backend as K
import tensorflow as tf

class ClusteringLayer(Layer):
    def __init__(self, n_clusters, **kwargs):
        super(ClusteringLayer, self).__init__(**kwargs)
        self.n_clusters = n_clusters
        self._initial_weights = None

    def build(self, input_shape):
        input_dim = input_shape[-1]  # This is where your error was
        self.clusters = self.add_weight(
            shape=(self.n_clusters, input_dim),
            initializer='glorot_uniform',
            name='clusters'
        )
        if self._initial_weights is not None:
            self.set_weights(self._initial_weights)
            self._initial_weights = None
        super().build(input_shape)

    def call(self, inputs):
        q = 1.0 / (1.0 + tf.reduce_sum(tf.square(tf.expand_dims(inputs, 1) - self.clusters), axis=2))
        q = tf.transpose(tf.transpose(q) / tf.reduce_sum(q, axis=1))
        return q

    def set_initial_weights(self, weights):
        self._initial_weights = [weights]


```

```python
def target_distribution(q):
    weight = q ** 1.5 / tf.reduce_sum(q, axis=0)
    return tf.transpose(tf.transpose(weight) / tf.reduce_sum(weight, axis=1))
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import SpectralClustering
import hdbscan

# Initialize output storage for silhouette scores
silhouette_scores = {
    'DEC (KMeans encoded)': [],
    'KMeans (encoded)': [],
    'KMeans': [],
    'Hierarchical': [],
    'GMM': [],
    'GMM (encoded)': [],
}

calinski_scores = {
    'DEC (KMeans encoded)': [],
    'KMeans (encoded)': [],
    'KMeans': [],
    'Hierarchical': [],
    'GMM': [],
    'GMM (encoded)': [],
}

davies_scores = {
    'DEC (KMeans encoded)': [],
    'KMeans (encoded)': [],
    'KMeans': [],
    'Hierarchical': [],
    'GMM': [],
    'GMM (encoded)': [],
}

dec_scores = {
    'q': [],
    'p': [],
    'y_pred': []
}

# Range of cluster numbers for classical methods
cluster_range = range(3, 13)

# For each number of clusters in the range
for n_clusters in cluster_range:

    encoded_x = encoder_model.predict(cities_clean_scaled)

    # KMeans on the encoded space
    kmeans = KMeans(n_clusters=n_clusters, n_init=20)
    y_pred_kmeans = kmeans.fit_predict(encoded_x)
    silhouette_scores['KMeans (encoded)'].append(silhouette_score(encoded_x, y_pred_kmeans))
    calinski_scores['KMeans (encoded)'].append(calinski_harabasz_score(encoded_x, y_pred_kmeans))
    davies_scores['KMeans (encoded)'].append(davies_bouldin_score(encoded_x, y_pred_kmeans))
    print(f"""Kmeans encoded computed for {n_clusters} clusters:
            silhouette            {silhouette_scores['KMeans (encoded)'][-1]}
            davies                {davies_scores['KMeans (encoded)'][-1]}
            calinski              {calinski_scores['KMeans (encoded)'][-1]}""")

    # KMeans on the original data space
    y_pred_kmeans_orig = kmeans.fit_predict(cities_clean_scaled)
    silhouette_scores['KMeans'].append(silhouette_score(cities_clean_scaled, y_pred_kmeans_orig))
    calinski_scores['KMeans'].append(calinski_harabasz_score(cities_clean_scaled, y_pred_kmeans_orig))
    davies_scores['KMeans'].append(davies_bouldin_score(cities_clean_scaled, y_pred_kmeans_orig))
    print(f"""Kmeans computed for {n_clusters} clusters:
            silhouette            {silhouette_scores['KMeans'][-1]}
            davies                {davies_scores['KMeans'][-1]}
            calinski              {calinski_scores['KMeans'][-1]}""")

    # Hierarchical clustering (Agglomerative)
    hierarchical = AgglomerativeClustering(n_clusters=n_clusters)
    y_pred_hierarchical = hierarchical.fit_predict(cities_clean_scaled)
    silhouette_scores['Hierarchical'].append(silhouette_score(cities_clean_scaled, y_pred_hierarchical))
    calinski_scores['Hierarchical'].append(calinski_harabasz_score(cities_clean_scaled, y_pred_hierarchical))
    davies_scores['Hierarchical'].append(davies_bouldin_score(cities_clean_scaled, y_pred_hierarchical))
    print(f"""Hierarchical encoded computed for {n_clusters} clusters:
            silhouette            {silhouette_scores['Hierarchical'][-1]}
            davies                {davies_scores['Hierarchical'][-1]}
            calinski              {calinski_scores['Hierarchical'][-1]}""")
    
    # GMM (Gaussian Mixture Model)
    gmm = GaussianMixture(n_components=n_clusters)
    y_pred_gmm = gmm.fit_predict(cities_clean_scaled)
    silhouette_scores['GMM'].append(silhouette_score(cities_clean_scaled, y_pred_gmm))
    calinski_scores['GMM'].append(calinski_harabasz_score(cities_clean_scaled, y_pred_gmm))
    davies_scores['GMM'].append(davies_bouldin_score(cities_clean_scaled, y_pred_gmm))
    print(f"""GMM computed for {n_clusters} clusters:
            silhouette            {silhouette_scores['GMM'][-1]}
            davies                {davies_scores['GMM'][-1]}
            calinski              {calinski_scores['GMM'][-1]}""")

    # GMM on the encoded space
    y_pred_gmm_encoded = gmm.fit_predict(encoded_x)
    silhouette_scores['GMM (encoded)'].append(silhouette_score(encoded_x, y_pred_gmm_encoded))
    calinski_scores['GMM (encoded)'].append(calinski_harabasz_score(encoded_x, y_pred_gmm_encoded))
    davies_scores['GMM (encoded)'].append(davies_bouldin_score(encoded_x, y_pred_gmm_encoded))
    print(f"""GMM encoded computed for {n_clusters} clusters:
            silhouette            {silhouette_scores['GMM (encoded)'][-1]}
            davies                {davies_scores['GMM (encoded)'][-1]}
            calinski              {calinski_scores['GMM (encoded)'][-1]}""")
    
    # --- Now DEC Model ---

    kmeans = KMeans(n_clusters=n_clusters).fit(encoded_x)

    # Build DEC model for the current number of clusters
    clustering_layer = ClusteringLayer(n_clusters=n_clusters, name='clustering')
    clustering_layer.set_initial_weights(kmeans.cluster_centers_)

    input_layer = Input(shape=(cities_clean_scaled.shape[1],))
    latent_output = encoder_model(input_layer)
    clustering_output = clustering_layer(latent_output)

    dec_model = Model(inputs=input_layer, outputs=clustering_output)

    # Compile model with dummy loss (handled manually)
    optimizer = SGD(learning_rate=0.01, momentum=0.9)
    loss_fn = tf.keras.losses.KLD

    # DEC training loop
    maxiter = 500
    update_interval = 5
    tol = 0.001
    batch_size = 32
    prev_sil = -1

    # Initial soft labels
    q = dec_model.predict(cities_clean_scaled, verbose=0)
    p = target_distribution(q)
    y_pred_last = q.argmax(1)

    for ite in range(maxiter):
        if ite % update_interval == 0:
            q = dec_model.predict(cities_clean_scaled, verbose=0)
            dec_scores["q"].append(q)
            p = target_distribution(q)
            dec_scores["p"].append(p)
            y_pred = q.argmax(1)

            # Evaluate silhouette
            latent_features = encoder_model.predict(cities_clean_scaled, verbose=0)
            sil = silhouette_score(latent_features, y_pred)
            print(f"Iter {ite}: Silhouette Score = {sil:.5f}")

            # Check convergence based on label stability
            delta_label = np.sum(y_pred != y_pred_last).astype(np.float32) / y_pred.shape[0]
            print(f"Label delta: {delta_label:.5f}")

            # Early stop if silhouette is dropping significantly
            if prev_sil != -1 and sil < prev_sil - tol:
                print("Stopping early due to silhouette drop")
                break
        
            if delta_label < tol:
                print(f"Converged at iteration {ite}")
                break
            y_pred_last = np.copy(y_pred)

        # Train one mini-batch step
        dec_model.train_on_batch(cities_clean_scaled, p)

        # Optional: periodic KL loss printing
        if ite % 100 == 0:
            loss = dec_model.evaluate(cities_clean_scaled, p, verbose=0)
            # print(f"Iter {ite}: KL loss = {loss:.5f}")

    # Store silhouette score for DEC model (after training)
    silhouette_scores['DEC (KMeans encoded)'].append(silhouette_score(encoded_x, y_pred_last))
    calinski_scores['DEC (KMeans encoded)'].append(calinski_harabasz_score(encoded_x, y_pred_last))
    davies_scores['DEC (KMeans encoded)'].append(davies_bouldin_score(encoded_x, y_pred_last))
    print(f"""DEC KMeans encoded computed for {n_clusters} clusters:
            silhouette            {silhouette_scores['DEC (KMeans encoded)'][-1]}
            davies                {davies_scores['DEC (KMeans encoded)'][-1]}
            calinski              {calinski_scores['DEC (KMeans encoded)'][-1]}""")
    dec_scores['y_pred'].append(y_pred_last+1)

# Output the final dictionary with all silhouette scores
print(silhouette_scores)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import pairwise_distances

# Initialize a dictionary to store the most representative cities_clean in each cluster
representative_cities_clean = {}

n_representatives = 5  # Specify the number of representative cities_clean per cluster
n_clusters = 10

rowname = ['y_pred_dec_' + str(n_clusters).zfill(2)]
print(rowname)
cluster_indices = np.where(cities_clean_sub[rowname] == cluster)[0]
cluster_data = cities_clean_scaled[cluster_indices]

print(len(cluster_data))

# Calculate the most representative cities_clean in each cluster
for cluster in range(1, n_clusters+1):

    # Calculate pairwise distances within the cluster
    distances = pairwise_distances(cluster_data)
    
    # Compute the average distance for each city to all other cities_clean in the cluster
    avg_distances = distances.mean(axis=1)

    print("avg dist computed")

    # Find the indices of the n_representatives cities_clean with the smallest average distance
    most_representative_indices = cluster_indices[np.argsort(avg_distances)[:n_representatives]]
    
    # Add the most representative cities_clean to the dictionary
    for i, index in enumerate(most_representative_indices):
        representative_city = cities_clean.iloc[index].copy()
        representative_city['Cluster'] = cluster
        representative_city['GHS_urban_area_id'] = cities_clean.iloc[index]['GHS_urban_area_id']
        # Use a unique key for each representative city
        key = f"{cluster}_{i+1}"
        representative_cities_clean[key] = representative_city
    
        # print("Cluster representative city id:", cities_clean.iloc[index]['GHS_urban_area_id'])

# Convert the dictionary to a DataFrame for better readability
representative_cities_clean_df = pd.DataFrame.from_dict(representative_cities_clean, orient='index')

representative_cities_clean_df["representative_city"] =  "representative"

print(representative_cities_clean_df[["representative_city", "GHS_urban_area_id"]])
```
