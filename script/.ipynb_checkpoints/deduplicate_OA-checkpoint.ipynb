```python
import pandas as pd

# Adjust the folder path as needed
dir = "/Users/simon/Documents/repo/cities-learning"

# Read files 1 through 10
dfs = []
for i in range(1, 11):
    filename = f"{dir}/data/OpenAlex/city_works_with_authors_df_{i}.feather"
    df = pd.read_feather(filename)
    dfs.append(df)

# Optionally concatenate them into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

print(combined_df.head())
```

```python
from semhash import SemHash

# Filter out rows with missing title or authors
filtered_df = combined_df.dropna(subset=["title", "authors"])

# Optionally convert to string if there's mixed types
filtered_df["title"] = filtered_df["title"].astype(str)
filtered_df["authors"] = filtered_df["authors"].astype(str)

# Convert to list of dictionaries
records = filtered_df[["title", "authors"]].to_dict(orient="records")

print(records)

# Deduplicate using SemHash
semhash = SemHash.from_records(records=records, columns=["title", "authors"])
deduplicated_records = semhash.self_deduplicate().deduplicated

# Convert back to a DataFrame
deduplicated_df = pd.DataFrame(deduplicated_records)
```

```python
# Number of rows before and after deduplication
original_count = len(filtered_df)
deduplicated_count = len(deduplicated_df)

print(f"Original (after dropna): {original_count}")
print(f"Deduplicated: {deduplicated_count}")
print(f"Removed during deduplication: {original_count - deduplicated_count}")
print(f"Percentage removed: {(original_count - deduplicated_count) / original_count:.2%}")

# Missing values in the original filtered dataframe
print("\nMissing values in original filtered DataFrame:")
print(filtered_df[["title", "authors"]].isna().sum())

# Missing values in the deduplicated dataframe
print("\nMissing values in deduplicated DataFrame:")
print(deduplicated_df[["title", "authors"]].isna().sum())

# Optional: Check for unique titles/authors
print("\nUnique counts:")
print("Original unique titles:", filtered_df["title"].nunique())
print("Deduplicated unique titles:", deduplicated_df["title"].nunique())
print("Original unique authors:", filtered_df["authors"].nunique())
print("Deduplicated unique authors:", deduplicated_df["authors"].nunique())
```

```python
from semhash import SemHash

thresholds = [0.75, 0.80, 0.85, 0.90, 0.95, 1]
results = []

for t in thresholds:
    semhash = SemHash.from_records(records=records, columns=["title", "authors"])
    dedup = semhash.self_deduplicate(threshold=t).deduplicated
    results.append((t, len(records), len(dedup), len(records) - len(dedup)))

# Print summary
print("\nThreshold | Original | Deduplicated | Removed")
for t, orig, dedup, removed in results:
    print(f"{t:<9} | {orig:<8} | {dedup:<13} | {removed}")
```

```python
import matplotlib.pyplot as plt

thresholds = [r[0] for r in results]
dedup_counts = [r[2] for r in results]

plt.plot(thresholds, dedup_counts, marker='o')
plt.title("Deduplicated Record Count vs. Similarity Threshold")
plt.xlabel("Similarity Threshold")
plt.ylabel("Remaining Records")
plt.grid(True)
plt.show()
```

```python
threshold = 0.9
semhash = SemHash.from_records(records=records, columns=["title", "authors"])
dedup_result = semhash.self_deduplicate(threshold=threshold)

removed = dedup_result.removed  # These are the ones considered duplicates

# Show a few examples
for i, r in enumerate(removed[:5]):
    print(f"\nDuplicate {i+1}:")
    print("Title:  ", r["title"])
    print("Authors:", r["authors"])
```
